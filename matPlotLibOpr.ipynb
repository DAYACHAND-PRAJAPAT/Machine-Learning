```python
# Data visualisation -> graphs and plot
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as ans
```

```python
arr1 = np.random.randint(2000,50000,100)
arr1
```

```python
plt.plot(arr1)
plt.show()
```

```python
arr1.sort()
plt.plot(arr1)
```

```python
x = [1,2,3,4,5,6]
y = [1,4,9,16,25,36]
plt.plot(x,y)
plt.plot(x,y, marker = 'H',
            ms = 15,
            mfc = "skyblue",
            mec = "black",
            color = 'red',
            linewidth = '3',
            ls = 'dashed',
            alpha = 0.5)
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
```

```python
## scatter plot
x = [1,2,3,4,5,6]
y = [1,4,9,16,25,36]
plt.scatter(x,y, marker='*')
```

```python
df = pd.read_csv('Used_Bikes.csv')
df.head()
```

```python
plt.show()
plt.scatter(df["price"],df['kms_driven'])
```

```python
## bar plot
x = ["Mp","Up","Rj","Mh"]
y = [3000,5000,4000,4500]
colours = ["red","green","blue","yellow"]
plt.bar(x,y,color=colours,
        width=0.3,
        edgecolor = 'black',
        label = x)
plt.legend()
plt.grid()
```

```python
## histogram -> distribution of data in bins(interval)
data = np.random.randn(1000)
plt.hist(data,bins=30)
plt.show()
```

```python
## pychart -> distribution of data in slices

labels = ["a","b","c","d"]
sizes = [10,20,30,40]
explodes = (0,0,0,0.07)
plt.pie(sizes,labels=labels,autopct="%1.2f%%",shadow=True,explode=explodes)
plt.title('Pie - Chart')
plt.savefig('plot.png')
plt.show()
```
